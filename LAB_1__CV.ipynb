# *TASK 1 ▶*
# *GROCERY LIST MANAGEMENT*

```python
grocery_list=[]

def add_item():
    item=input("Enter the item to add : ")
    grocery_list.append(item)
    print(f"{item} has been added to the list.")

def remove_item():
    item=input("Enter the item to remove : ")
    if item in grocery_list:
        grocery_list.remove(item)
        print(f"{item} has been removed from the list.")
    else:
        print(f"{item} not found in the list.")

def display_list():
    if not grocery_list:
        print("Grocery list is empty !")
    else:
        print("\n -- YOUR GROCERY LIST --")
        for i, item in enumerate(grocery_list, 1):
            print(f"{i}. {item}")

while True:
    print("\n--- GROCERY LIST MENU ---")
    print("1. Add item")
    print("2. Remove item")
    print("3. Display list")
    print("4. Exit")
    choice=input("Enter your choice (1-4) : ")
    if choice == "1":
        add_item()
    elif choice == "2":
        remove_item()
    elif choice == "3":
        display_list()
    elif choice == "4":
        print("EXIT ...")
        break
    else:
        print("INVALID CHOICE! Please enter 1–4.")
```

# *TASK 2 ▶*
# *STUDENT RECORD SYSTEM*

```python
student_rec={}

def add_student():
    student_id=input("Enter student ID: ")
    if student_id in student_rec:
        print("Student ID already exists !")
    else:
        name=input("Enter student name : ")
        grade=input("Enter student grade : ")
        student_rec[student_id]=[name, grade]
        print(f"Student {name} added successfully.")

def update_grade():
    student_id=input("Enter student ID to update grade : ")
    if student_id in student_rec:
        new_grade=input("Enter new grade : ")
        student_rec[student_id][1]=new_grade
        print(f"Grade updated for {student_rec[student_id][0]}.")
    else:
        print("STUDENT NOT FOUND !")

def display_records():
    if not student_rec:
        print("NO STUDENT RECORDS AVAILABLE !")
    else:
        print("\n--- STUDENT RECORDS  ---")
        for sid, data in student_rec.items():
            print(f"ID: {sid}, NAME: {data[0]}, GRADE: {data[1]}")

while True:
    print("\n--- STUDENT RECORD SYSTEM MENU ---")
    print("1. Add Student")
    print("2. Update Grade")
    print("3. Display Records")
    print("4. Exit")
    choice=input("Enter your choice (1-4): ")
    if choice == "1":
        add_student()
    elif choice == "2":
        update_grade()
    elif choice == "3":
        display_records()
    elif choice == "4":
        print("EXIT ...")
        break
    else:
        print("INVALID CHOICE ! Please enter 1–4.")
```

# *TASK 5 ▶*

```python
import cv2
import matplotlib.pyplot as plt

image=cv2.imread("/content/P3.jpg")
if image is None:
    print("Error : Image not found!")
else:
    image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    blurred=cv2.GaussianBlur(image,(25,25),0)
    blurred_rgb=cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB)
    h, w, _ =image.shape
    ch, cw =600,600
    start_y=h//2-ch//2
    start_x=w//2-cw//2
    cropped=image_rgb[start_y:start_y+ch, start_x:start_x+cw]

    plt.imshow(image_rgb)
    plt.title(" -- ORIGINAL IMAGE --")
    plt.axis("off")
    plt.show()

    plt.imshow(blurred_rgb)
    plt.title(" -- GAUSSIAN BLURRED IMAGE --")
    plt.axis("off")
    plt.show()

    plt.imshow(cropped)
    plt.title(" -- CROPPED IMAGE --")
    plt.axis("off")
    plt.show()
```

# *TASK 7 ▶*

```python
import cv2
import matplotlib.pyplot as plt

image=cv2.imread("/content/p2.jpg", cv2.IMREAD_GRAYSCALE)

if image is None:
    print("Error: Image not found!")
else:
    # pixel between 127 to 255 (white) else 0 (black)
    _, binary = cv2.threshold(image,127,255,cv2.THRESH_BINARY)
    h, w=binary.shape
    center=(w // 2, h // 2)
    #  parametres for rotation (center, angle, scale)
    rotation_matrix=cv2.getRotationMatrix2D(center,60,1.0)
    rotated=cv2.warpAffine(binary, rotation_matrix, (w, h))

    plt.imshow(image,cmap="gray")
    plt.title(" -- ORIGINAL GRAYSCALE -- ")
    plt.axis("off")
    plt.show()

    plt.imshow(binary, cmap="gray")
    plt.title(" -- BINARY THRESHOLD --")
    plt.axis("off")
    plt.show()

    plt.imshow(rotated, cmap="gray")
    plt.title("-- ROTATION 60 DEGREES --")
    plt.axis("off")
    plt.show()
```

# *TASK 8 ▶*

```python
import cv2
import matplotlib.pyplot as plt

img1=cv2.imread("/content/P5.jpg")
img2=cv2.imread("/content/P6.jpg")

if img1 is None or img2 is None:
    print("Error: One or both images not found!")
else:
    # both images should be of same size to blend
    img2=cv2.resize(img2,(img1.shape[1],img1.shape[0]))
    # alpha=0.5 means 50-50 blend we can adjust blend accordingly
    blended=cv2.addWeighted(img1, 0.3, img2, 0.5, 0)

    gray=cv2.cvtColor(blended,cv2.COLOR_BGR2GRAY)
    equalized=cv2.equalizeHist(gray)

    img1_rgb=cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
    img2_rgb=cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
    blended_rgb=cv2.cvtColor(blended, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=(10,10))
    plt.subplot(2,3,1)
    plt.imshow(img1_rgb)
    plt.title(" -- IMAGE 1 --")
    plt.axis("off")

    plt.figure(figsize=(10,10))
    plt.subplot(2,3,2)
    plt.imshow(img2_rgb)
    plt.title(" -- IMAGE 2 --")
    plt.axis("off")

    plt.figure(figsize=(10,10))
    plt.subplot(2,3,3)
    plt.imshow(blended_rgb)
    plt.title(" -- BLENDED IMAGE --")
    plt.axis("off")

    plt.figure(figsize=(10,10))
    plt.subplot(2,3,4)
    plt.imshow(gray, cmap="gray")
    plt.title("-- GRAYSCALE IMAGE --")
    plt.axis("off")

    plt.figure(figsize=(10,10))
    plt.subplot(2,3,5)
    plt.imshow(equalized, cmap="gray")
    plt.title("-- HISTOGRAM EQUALIZED --")
    plt.axis("off")
    plt.show()
```

# *TASK 9 ▶*

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img1=np.zeros((300, 300), dtype=np.uint8)
img2 = np.zeros((300, 300), dtype=np.uint8)

cv2.rectangle(img1, (50, 50), (150, 150), 255, -1)
cv2.rectangle(img2, (100, 100), (200, 200), 255, -1)

# bitwise operations
bitwise_and=cv2.bitwise_and(img1, img2)
bitwise_or=cv2.bitwise_or(img1, img2)
bitwise_xor=cv2.bitwise_xor(img1, img2)
bitwise_not1=cv2.bitwise_not(img1)
bitwise_not2=cv2.bitwise_not(img2)


titles=['-- Image 1 --',' -- Image 2 --','-- AND --','-- OR --','-- XOR --','-- NOT (Image 1) --','-- NOT (Image 2) --']
images=[img1,img2,bitwise_and,bitwise_or,bitwise_xor,bitwise_not1,bitwise_not2]

plt.figure(figsize=(10,8))
for i in range(len(images)):
    plt.subplot(3,3,i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```

# *TASK 10 ▶*

```python
import cv2
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

img=cv2.imread("/content/P6.jpg")
img_rgb=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# image to Pandas df
pixels=img_rgb.reshape(-1, 3)
df=pd.DataFrame(pixels, columns=["Red", "Green", "Blue"])

print("\n       -- STATISTICS OF PIXEL VALUES -- \n ")
print(df.describe())

# circle mask
mask=np.zeros(img.shape[:2], dtype=np.uint8)
cv2.circle(mask,(img.shape[1]//2, img.shape[0]//2), 100, 255, -1)

# bitwise AND with mask
masked_img=cv2.bitwise_and(img_rgb, img_rgb, mask=mask)

plt.figure(figsize=(12,6))

plt.subplot(1,3,1)
plt.imshow(img_rgb)
plt.title(" -- Original Image --")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(mask, cmap="gray")
plt.title(" -- Circle Mask --")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(masked_img)
plt.title(" -- Masked Image --")
plt.axis("off")

plt.tight_layout()
plt.show()
```
